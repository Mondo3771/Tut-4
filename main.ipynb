```python
import random
import math as m
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
    return 1 / (1 + m.exp(-x))
# Define the function f(x)
def f(x):
    return (x**2)*m.sin(2*m.pi*x) + 0.7

# Generate 100 points and their classes
points = [(random.uniform(0, 1), random.uniform(0, 1)) for _ in range(100)]
classes = [0 if f(x1) > x2 else 1 for x1, x2 in points]
# Separate the points by class
class_0 = [point for point, class_label in zip(points, classes) if class_label == 0]
class_1 = [point for point, class_label in zip(points, classes) if class_label == 1]

# Plot the points with different symbols for the two classes
plt.scatter(*zip(*class_0), marker='o', label='Class 0')
plt.scatter(*zip(*class_1), marker='x', label='Class 1')

# Set the title and labels
plt.title('Random Points and Classes')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
points = np.array(points)
ones = np.ones((points.shape[0], 1))  # Create a column of 1s
points = np.hstack((points, ones))  # Add the column of 1s to 'points'

print(points)
# Show the plot
```
